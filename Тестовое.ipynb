**Задание 1**  

Города A, B, C вместе с соединяющими их прямыми дорогами образуют треугольник.
Известно, что прямой путь из А в B на 200 км короче, чем объезд через C, а прямой
путь из A в C на 300 км короче, чем объезд через B. Найдите расстояние между
городами B и С


---
У нас имеется треугольник ABC, нам нужно найти его сторону BC.

<img src="https://avatars.mds.yandex.net/get-images-cbir/4010443/Z7UOs6l5W4Aic1YUtmVQXQ1932/ocr" width="255" height="200">

Обозначим стороны треугольника:
\begin{equation}
AB = x, BC = z, AC = y
\end{equation}
По условию
\begin{equation}
ACB - AB = 200
\end{equation}
\begin{equation}
ABC - AC = 300
\end{equation}
\begin{equation}
=> ACB = y + z, ABC = x + z
\end{equation}
Тогда получим систему из 2-х уравнений:
\begin{cases}
   y + z - x = 200 \\
   x + z - y = 300
 \end{cases}
Прибавив к первому уравнению второе, получим:
\begin{equation}
y + z - x + x + z - y = 200 + 300
\end{equation}
В итоге получаем, что
\begin{equation}
BC = z = 500 / 2 = 250
\end{equation}

Ответ: BC = 250






**Задание 2**

Компания ПЭК тестирует доставку последней мили с помощью автономных
роботов-доставщиков. На сортировочном центре в черте города роботу загружают
посылку (помещается только одна), далее робот отправляется по адресу, где его
встречает получатель и забирает посылку. Для доставки следующей посылки робот
возвращается на сортировочный центр.
Сортировочный центр обслуживает район города, очерченный окружностью радиуса R.
Для простоты расчетов будем считать, что клиенты по этой площади распределены
равномерно, а робот двигается к клиенту вдоль радиуса. Скорость движения робота
постоянна = v. Временем, необходимым на загрузку и разгрузку посылки, можно
пренебречь.
Как связано количество посылок, которое успеет доставить робот за день и
предельная дальность доставки R?

Сделайте оценку среднего количества доставок за день с указанием всех значимых констант. Приложите не только ответ, но и ход решения.


---

<img src="https://avatars.mds.yandex.net/get-images-cbir/1599092/MGILgBEWgkI5MamzFfrajQ1798/ocr" width="300" height="255">

Пусть среднего количества доставок за день z.

Для начала перечислим все значимые константы:
\begin{equation}
v - скорость\ движения\ робота, \\
R - радиус\ окружности\ района,\ центром\ которого\ является\ сортировочный\ центр, \\
T - количество\ времени\ в\ день,\ в\ течение\ которого\ робот\ может\ доставлять\ посылки.
\end{equation}

По условию нам не важно, в каком направлении движется робот от центра (сервисного центра - СЦ) к любой точке внутри окружности (адрес получателя посылки), а важно лишь расстояние, на котором эта точка находится от центра. Следовательно, мы имеем набор расстояний
\begin{equation}
a_i \in [{0,R}],
\end{equation}
распредленных равномерно (так как клиенты распределены по площади равномерно). Так как мы учитываем не только путь от СЦ до адреса, но и обратный путь, то нам понадобится набор удвоенных расстояний \a_i:
\begin{equation}
b_i = 2 * a_i \\
=> b_i \in [{0,2R}],
\end{equation}
которые также распределены равномерно с функцией распределения
\begin{equation}
f(x) =
\begin{cases}
   \frac{1}{2R},\ 0 \leqslant x \leqslant 2R\\
   0, x < 0\ или\ x>2R
 \end{cases}
\end{equation}
Найдем среднее расстояние \b_i, которое будет проходить робот от СЦ до адреса и обратно, как матожидание случайно величины, распрделенной равномерно:
\begin{equation}
b_{ср} =  \int_{0}^{2R} x f(x)\ dx = \int_{0}^{2R} x \frac{1}{2R} dx = \frac{1}{2R} \int_{0}^{2R} x dx = \\ = \frac{1}{2R} \frac{(2R)^2}{2} = R
\end{equation}
Тогда получается, что среднее время, затрачиваемое на доставку одной посылки, равно:
\begin{equation}
t_{ср} = \frac{b_{ср}}{v} = \frac{R}{v}
\end{equation}
С учетом того, что время работы робота в день равно T, получаем среднее число доставленных посылок за день:
\begin{equation}
z = \frac{T}{t_{ср}} = \frac{T}{\frac{R}{v}} = \frac{Tv}{R}
\end{equation}

Ответ: z = Tv/R
